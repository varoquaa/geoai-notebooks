# Get metadata and clip raster by bounding box using geoai

## Import geoai

```python
import geoai
```

## Set raster and vector urls for download

```python
raster_url = ('https://huggingface.co/datasets/giswqs/geospatial/resolve/main/naip_test_buildings.tif')

vector_url = ('https://huggingface.co/datasets/giswqs/geospatial/resolve/main/naip_test_buildings.geojson')
```

## Download raster

```python
raster_path = geoai.download_file(raster_url)
```

## Download vector

```python
vector_path = geoai.download_file(vector_url)
```

## Get Metadata

```python
geoai.print_raster_info(raster_path)
```

## Get vector Metadata

```python
geoai.print_vector_info(vector_path)    
```

## Clip the raster by a Bounding Box

```python
clip_raster_path = "naip_clip.tif"

geoai.clip_raster_by_bbox(raster_path, clip_raster_path, bbox=(0, 0, 1000, 1000))

geoai.view_image(clip_raster_path)
```
